```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def q(b,m):
    eps = 8; delf = 2/(m+1)
    x = math.exp(eps/b); y = math.exp(eps); z = y**m
    return (x/(2*((x+1)**2)*(y-1)))*((y-1)*(y*z-1)-delf*(y/x)*(m*y*z-(m+1)*z+1))
            
def X(b,m):
    eps = 8
    x = math.exp(eps/b); y = math.exp(eps); z = y**m
    return (y*z)/(2*q(b,m)*(x+1))

def Ys(b,m):
    eps = 8; delf = 2/(m+1)
    Y = np.zeros(m+2)
    x = math.exp(eps/b); y = math.exp(eps); z = y**m
    for j in range(1,m+2):
        Y[j] = (x/(2*q(b,m)*(x+1)))*(1+delf*(y/x)*((m+1-j)*z-(z-1)/(y-1)))
    return Y
```

```python
def V0(b,m):
    eps = 8
    qq = q(b,m); XX = X(b,m); YY = Ys(b,m); DD = XX + YY[1]
    y = math.exp(eps)
    T = 0
    for j in range(1,m+2):
        T += (YY[j]**3)/(y**(m+1-j))
    return (2*qq/3)*((y-1)/y)*T+(2*qq*(DD**3))/(3*(y**(m+1)))

def V1(b,m):
    eps = 8
    qq = q(b,m); XX = X(b,m); YY = Ys(b,m)
    y = math.exp(eps)
    S = 0
    for j in range(1,m+2):
        S += YY[j]/(y**(m+1-j))
    return 2*qq*(XX**2)*((y-1)/y)*S-1+V0(b,m)
```

```python
m = 0

eps = 8; delf = 2/(m+1)
y = math.exp(eps)
A = delf*(y*(y**m-1)/(y-1)); B = (1+m*delf)*(y**(m+1))
U = eps/(math.log((A+math.sqrt(A**2+4*B))/2))
fig = plt.figure()
#plt.ylim(0.6, 1.25)
plt.xlim(-0.05, 1.05)
plt.plot([0,1], [V0(U,m),V1(U,m)], color='black', linestyle='dashed', label='$b = U$')
plt.plot([0,1], [V0((m+3)*U/(m+4),m),V1((m+3)*U/(m+4),m)], color='red', linestyle='-', label=r"$b = \frac{m+3}{m+4} \cdot U$")
plt.plot([0,1], [V0((2*m+7)*U/(2*(m+4)),m),V1((2*m+7)*U/(2*(m+4)),m)], color='blue', linestyle='dotted', label=r"$b = \frac{2m+7}{2(m+4)} \cdot U$")
plt.plot([0,1], [V0((2*m+5)*U/(2*(m+4)),m),V1((2*m+5)*U/(2*(m+4)),m)], color='green', linestyle='dashdot', label=r"$b = \frac{2m+5}{2(m+4)} \cdot U$")
plt.legend(bbox_to_anchor=(0.49, 1.02), loc='lower center', borderaxespad=0, ncol=2, fontsize=13.5)
plt.xlabel("$t^2$", fontsize = 18)
plt.ylabel("$Var[t^*]$", fontsize = 18)
fig.savefig("Variance_CPM_m=0_epsilon=8.png")
fig.savefig("figs/Variance_CPM_m=0_epsilon=8.eps", bbox_inches="tight", pad_inches=0.05)
```
